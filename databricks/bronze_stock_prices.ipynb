```python
df = (
    spark.read
    .option("header", True)
    .option("inferSchema", True)
    .csv("/Volumes/workspace/default/stock/raw")
)
```

```python
df.write \
  .format("delta") \
  .mode("overwrite") \
  .saveAsTable("bronze_stock_prices")
```

```python
print("Bronze table row count:")
spark.table("bronze_stock_prices").count()
```

```python
%sql
-- creating bronze_portfolio_transactions using sql
CREATE OR REPLACE TABLE bronze_portfolio_transactions
USING DELTA
AS
SELECT
    stock,
    transaction_type,
    CAST(quantity AS DOUBLE) AS quantity,
    CAST(price AS DOUBLE) AS price,
    CAST(transaction_date AS DATE) AS transaction_date
FROM read_files(
    '/Volumes/workspace/default/stock/portfolio/portfolio_transactions.csv',
    format => 'csv',
    header => true
);
```
